```python
import pandas as pd

# Cargar datos

url='https://raw.githubusercontent.com/santiloc-hub/data_analitica2/refs/heads/main/bank_final.csv'
df = pd.read_csv(url)

# Verificar datos nulos
print(df.isnull().sum())

# Eliminar filas con datos nulos
#df.dropna(inplace=True)
df.head()
```

```python
# Crear una nueva columna binaria para facilitar el cálculo
df['y_binary'] = df['y'].apply(lambda x: 1 if x == 'yes' else 0)

# Agrupar por trabajo y educación, y contar el número de "sí" y "no"
yes_counts = df[df['y'] == 'yes'].groupby(['job', 'education']).agg(
    count_yes=('y_binary', 'size'),
    avg_balance_yes=('balance', 'mean')
).reset_index()

no_counts = df[df['y'] == 'no'].groupby(['job', 'education']).agg(
    count_no=('y_binary', 'size'),
    avg_balance_no=('balance', 'mean')
).reset_index()

# Unir los resultados en un solo DataFrame
merged_counts = pd.merge(yes_counts, no_counts, on=['job', 'education'], how='outer').fillna(0)

# Convertir los resultados en enteros donde corresponde
merged_counts['count_yes'] = merged_counts['count_yes'].astype(int)
merged_counts['count_no'] = merged_counts['count_no'].astype(int)

# Mostrar el resultado final con el cálculo del balance
print(merged_counts)
merged_counts['yes_probability']=(merged_counts['count_yes']/(merged_counts['count_no']+merged_counts['count_yes']))*100
merged_counts.head(50)
```
